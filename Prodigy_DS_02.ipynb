# Handling Missing Values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')

data[['Cabin']] = imputer.fit_transform(data[['Cabin']])
```

# Exploratory Data Analysis

```python
selected_data = data[['Survived', 'Age', 'Fare', 'Pclass']].dropna()

# Create a pair plot
sns.pairplot(selected_data, hue='Survived')
plt.show()
```

## Visualize Distribution by AGE

```python
sns.histplot(data['Age'].dropna(), kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Create a DataFrame with the counts
survival_counts = data.groupby(['Survived', 'Sex']).size().unstack()

# Plot pie charts
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

survival_counts.loc[0].plot.pie(ax=axes[0], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
axes[0].set_title('Did Not Survive')
axes[0].set_ylabel('')

survival_counts.loc[1].plot.pie(ax=axes[1], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
axes[1].set_title('Survived')
axes[1].set_ylabel('')

plt.suptitle('Survival Count by Gender')
plt.show()
```

# Visualize the survival rate by pasenger class

```python
sns.countplot(x='Survived', hue='Pclass', data=data, palette={1: 'red', 2: 'blue', 3: 'green'})
plt.title('Survival Count by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Embarked', data=data)
plt.title('Survival Count by Embarked Port')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```
